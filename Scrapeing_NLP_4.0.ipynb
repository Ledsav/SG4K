```python
#------------------------------------DESCRIPTION-------------------------------------------
#this function is used to retrive appids from the playstore based on the key words passed.
# lenguage = english
#you have to pass a list with the search words e.g ['bambini' ' '  'apprendimento']
#------------------------------------------------------------------------------------------



# aggiungi controllo su gioco approvato 


def retrive_data(search_words):
    
        import time
        from selenium import webdriver
        from webdriver_manager.chrome import ChromeDriverManager
#------------------ADDED----------------------------------------       
        lista=list()
        for word in search_words:
            lista.append(word)
            lista.append('%20')
        lista.pop()   
        lista = [''.join(lista)]  

        url = ['https://play.google.com/store/search?q=','','&c=apps&hl=us&gl=us']
        url[1] = lista[0]
        url = ''.join(url)
#--------------------------------------------------------------- 
        driver = webdriver.Chrome(ChromeDriverManager().install())
        driver.get(url)
        time.sleep(10) 


        SCROLL_PAUSE_TIME = 5 
        # Get scroll height
        last_height = driver.execute_script("return document.body.scrollHeight")

        time.sleep(SCROLL_PAUSE_TIME)
        links_games=[]
        while True:
                # Scroll down to bottom
                driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

                # Wait to load page
                time.sleep(SCROLL_PAUSE_TIME)

                # Calculate new scroll height and compare with last scroll height
                new_height = driver.execute_script("return document.body.scrollHeight")
                if new_height == last_height:
                    break
                last_height = new_height

                links_games = []
                elems = driver.find_elements_by_xpath("//a[@href]")
                for elem in elems:
                    if "details?id" in elem.get_attribute("href"):
                        links_games.append((elem.get_attribute("href")))

                links_games = list(dict.fromkeys(links_games))

                list_all_elements = []
        for iteration in links_games:
            try:
                driver.get(iteration)
                print(iteration)
                time.sleep(3)

                header1 = driver.find_element_by_tag_name("h1")
                star = driver.find_element_by_class_name("BHMmbe")


                others = driver.find_elements_by_class_name("htlgb")
                list_others = []
                for x in range (len(others)):
                    if x % 2 == 0:
                        list_others.append(others[x].text)

                titles = driver.find_elements_by_class_name("BgcNfc")
                comments = driver.find_element_by_class_name("EymY4b")

                list_elements = [iteration,header1.text, float(star.text.replace(",",".")),]


                list_all_elements.append(list_elements)
            except Exception as e:
                print(e)



        import pandas as pd
        for x in list_all_elements:
            x[0] = x[0][46:]

        df = pd.DataFrame(list_all_elements,columns=['URL', 'Name', 'Stars'])
        df.to_excel('scraping_playstore.xlsx', header = True, index=False)
```

```python
#tokenization for app infos

#------------------------------------DESCRIPTION-------------------------------------------
#this function is used to tokenize all the infromations about an app in a dictionary of type {'appid':'stemmed_text'}
#in addiction the metrics BoW and Tf_Idf are computed over the all dataset. 
# THIS SECOND PART IS VERY IMPORTANT IN THE PHASE OF VOCABULARY DEFINITION.
#------------------------------------------------------------------------------------------

def tokeinzation_Bow_Tf_Idf(dataframe):


        import nltk
        import numpy as np
        from nltk.tokenize import word_tokenize
        from nltk.corpus import stopwords
        from nltk.stem.porter import PorterStemmer
        from nltk.stem import WordNetLemmatizer 
        from nltk.probability import FreqDist
        import string
        stemmedLists=list()
        stemmedLists_dict=dict()
        lemmedLists=list()
        lemmedLists_dict=dict()
        BoW=list()
        BoW_dict= dict()
        app_infos_df=dataframe
        nltk.download('words')
        dic = set(nltk.corpus.words.words())

        for i in range(len(app_infos_df)):
            
            noise_free_text = app_infos_df.loc[i,'description'].split("<b>") 
            noise_free_text = ''.join(noise_free_text)
            noise_free_text = noise_free_text.split("</b>") 
            noise_free_text = ''.join(noise_free_text)
            tokens = word_tokenize(noise_free_text)
            tokens = [w.lower() for w in tokens]
            table = str.maketrans('', '', string.punctuation)
            stripped = [w.translate(table) for w in tokens]
            words = [word for word in stripped if word.isalpha() and word in dic]
            stop_words = stopwords.words('english')
            words = [w for w in words if not w in stop_words]
            
            
#-----------------------Stemmization-----------------------------            
            porter = PorterStemmer()
            stemmed = [porter.stem(word) for word in words]
#            stemmedLists.append(stemmed)
#            stemmedLists_dict[app_infos_df.loc[i,'title']] = stemmed
            
#-----------------------Lemmization------------------------------


            lemmatizer = WordNetLemmatizer()
            lemm = [lemmatizer.lemmatize(word) for word in stemmed if len(word)>2]
            
            lemmedLists.append(lemm)
            lemmedLists_dict[app_infos_df.loc[i,'title']] = lemm

#----------------------------------------------------------------
         #using lemmed
    
    
            BoW.append(FreqDist(lemm))
            BoW_dict[app_infos_df.loc[i,'title']] = dict(BoW[i])


        
        Tf_Idf_it=dict()
        Tf_Idf_dict = dict()
        j=0
        i=0
        count=list()
        count_k=list()
        number=list()
        number_list=list()
        for ii in lemmedLists:
            ii = set(ii)
            ii = list(ii)
            for m in ii:
                for k in lemmedLists:
                    count.append(k.count(m))
                count_k.append(count)
                count=list()
                number.append(np.count_nonzero(count_k[i]))
                i=i+1   
            number_list.append(number)
            number=list()   
        for i in range(len(BoW)):
             for elm in list(BoW[i].values()):
                    Tf_Idf_it[list(BoW[i].keys())[j]] = (len(BoW)/number_list[i][j])* (elm) #keap out normalization on length of description
                    j=j+1
             Tf_Idf_dict[app_infos_df.loc[i,'title']] =Tf_Idf_it
             j = 0
             Tf_Idf_it = dict()
        
        return lemmedLists_dict,BoW_dict,Tf_Idf_dict
  


```

```python
#function to tokenize text

def Tokenize_txt(list_Of_Words):
        import nltk
        import numpy as np
        from nltk.tokenize import word_tokenize
        from nltk.corpus import stopwords
        from nltk.stem.porter import PorterStemmer
        from nltk.stem import WordNetLemmatizer 
        import string
        lemmedLists=list()
        nltk.download('words')
        dic = set(nltk.corpus.words.words())
  
        tokens = word_tokenize(list_Of_Words)
        tokens = [w.lower() for w in tokens]
        words = [word for word in tokens if word.isalpha()and word in dic]
        stop_words = stopwords.words('english')
        words = [w for w in words if not w in stop_words]
        porter = PorterStemmer()
        stemmed = [porter.stem(word) for word in words]
        lemmatizer = WordNetLemmatizer()
        lemm = [lemmatizer.lemmatize(word) for word in stemmed]
        return lemm
```

```python
#Obtain the reviews 
import pandas as pd
res_tot_scraped_df = pd.read_csv('res_tot_scraped.csv')
app_infos_df = pd.read_csv('apps.csv')
prov = pd.merge(app_infos_df,res_tot_scraped_df [res_tot_scraped_df ['SG']=='YES'], on=['title'], how='inner')
app_ids= list(prov['appId'])
app_reviews = []

from tqdm import tqdm
from google_play_scraper import Sort, reviews, app

#takeing the most relevant and newest  
for ap in tqdm(app_ids):
    for score in list(range(1, 6)):
        for sort_order in [Sort.MOST_RELEVANT, Sort.NEWEST]:
            rvs, _ = reviews(
            ap,
            lang='en',
            country='us',
            sort=sort_order,
            count= 20,
            filter_score_with=score
          )
        for r in rvs:
            r['sortOrder'] = 'most_relevant' if sort_order == Sort.MOST_RELEVANT else 'newest'
            r['appId'] = ap
            app_reviews.extend(rvs)
            
            
            
app_reviews_df = pd.DataFrame(app_reviews)
app_reviews_df.to_csv('reviews.csv', index=None, header=True)            
```

```python
import tweepy
from textblob import TextBlob 
from wordcloud import WordCloud 
import numpy as np 
import re 
import matplotlib.pyplot as plt 
plt.style.use('fivethirtyeight')

# This functions define how much a comment is subjective and how much is positive------------------------------------------
def getSubj(text):
    return TextBlob(text).sentiment.subjectivity
#Polarity 
def getPol(text):
    return TextBlob(text).sentiment.polarity

# This function discretize the values of polarity--------------------------------------------------------------------------
def getAnalysis(score):
    if score <0:
        return 'NEGATIVE'
    elif score == 0:
        return 'NEUTRAL'
    else :
        return 'POSITIVE'

# This function tells the percentage of postivie, negative and neutral reviews---------------------------------------------
def getPercentage(pol,df_t):

    if pol == ('pos'):
        positiv = df_t[df_t.pol > 0]
        positiv = positiv['pol']

        prc = round((positiv.shape[0] /df_t.shape[0])*100,1)

    elif pol == ('neg'):
        positiv = df_t[df_t.pol < 0]
        positiv = positiv['pol']

        prc = round((positiv.shape[0] /df_t.shape[0])*100,1)



    elif pol == ('neu'):
        positiv = df_t[df_t.pol == 0]
        positiv = positiv['pol']

        prc = round((positiv.shape[0] /df_t.shape[0])*100,1)

    return prc



#---------------------------Constructing the dataset cleaning the reviews----------------------------------------

import numpy as np
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from nltk.probability import FreqDist
from nltk.stem import WordNetLemmatizer 
import string

lemmedLists=list()
rev = list()
rev_df=pd.DataFrame()
rev_df_mean=pd.DataFrame()
rev_df_mean_tot=pd.DataFrame()
posl=list()
negl=list()
rev_df_mean_tot_dict=dict()
i=0
j=0
app_reviews_df = pd.read_csv('reviews.csv')

for x in list(set(app_reviews_df.appId)):
        rev_df=pd.DataFrame()
        prov=  app_reviews_df[app_reviews_df.loc[:, ('appId')] == x]
        prov = prov[['content']]
        for k in prov[['content']].itertuples():
            if not pd.isna(k.content):
                tokens = word_tokenize(k.content)
                tokens = [w.lower() for w in tokens]
                table = str.maketrans('', '', string.punctuation)
                stripped = [w.translate(table) for w in tokens]
                words = [word for word in stripped if word.isalpha()]
                stop_words = stopwords.words('english')
                words = [w for w in words if not w in stop_words]
                porter = PorterStemmer()
                stemmed = [porter.stem(word) for word in words]
                lemmatizer = WordNetLemmatizer()
                lemm = [lemmatizer.lemmatize(word) for word in stemmed if len(word)>2]
                lemm_un =' '.join(lemm)
                sub = getSubj(lemm_un)
                pol = getPol(lemm_un)
                #--------computing the metrics for every review on appId k-----------------
                di = {'id_comment':i,'lemmized':lemm_un,'sub':sub,'pol':pol}
                rev_df = rev_df.append(di, ignore_index = True)
                i+=1
        #--------computing the mean of the meterics for k reviews-----------------        
        rev_df_mean= rev_df.mean(axis = 0)
        pos = getPercentage('pos',rev_df)
        neg = getPercentage('neg',rev_df)
        rev_df_mean_tot_dict[x]={'sub':rev_df_mean[1], 'pol':rev_df_mean[2],'perc_pos':pos,'perc_neg':neg}

```
